```python
import matplotlib.pyplot as plt
```

```python
def periodic_function(x):
    if (abs(x)<=0.5): 
        return x 
    
    decimal=x-np.floor(x) 
    if (decimal<=0.5): 
        return decimal 
    return decimal-1
    


x = np.linspace(-1, 1, 1000)  
y_vectorized=np.vectorize(periodic_function)
y = y_vectorized(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, label=r'$f(x) = x \, \mathrm{mod} \, 1$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Periodic Function $f(x) = x$ with Period 1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

```python
#  for terms -M to M 
M=10
L=1 
dx=0.001 
x = np.linspace(-L, L, np.int32(2*L/dx)+1, endpoint=True)
s=0
for n in np.linspace(-M, M, 2*M+1, endpoint=True): 
    an=0
    if (n!=0): 
        an=1j*np.cos(np.pi*n)/(2*np.pi*n)+1j*np.sin(np.pi*n)/(2*np.pi*np.pi*n*n)  
    s+=an*np.exp(1j*2*np.pi*n*x) 
plt.figure(1)
plt.plot(x, np.real(s))  
plt.title("Fourier Series for "+str(-1*M)+" to "+str(M)+" terms") 
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show() 
```

```python
def rect(x, a): 
    return np.where(np.abs(x) <= a, 1.0, 0.0)
    
def f(x, y, a, b):
    return rect(x, a)*rect(y, b)

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)


z = f(x, y, 2, 4)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(x, y, z, cmap='viridis')


ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis (G(f_x, f_y))')


plt.show()



```

```python
from mpl_toolkits.mplot3d import Axes3D

def G(f_x, f_y, a, b):
    return 4*a*b*np.sinc(2*b*f_x)*np.sinc(2*a*f_y)


x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
x, y = np.meshgrid(x, y)


z = G(x, y, 2, 4)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.plot_surface(x, y, z, cmap='viridis')


ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis (G(f_x, f_y))')

plt.show()
```
